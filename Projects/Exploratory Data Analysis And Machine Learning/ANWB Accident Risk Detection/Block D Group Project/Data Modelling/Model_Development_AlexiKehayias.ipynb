```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import combinations
import numpy as np
 
# Assuming joined_df is your DataFrame containing the data
# Select relevant columns
columns = [ 'event_start', 'event_end',
           'duration_seconds', 'longitude', 'speed_kmh', 'end_speed_kmh',
           'maxwaarde', 'category', 'incident_severity', 'is_valid',
            'road_manager_type', 'road_number', 'road_name', 'place_name',
           'municipality_name', 'road_manager_name']
 
# Ensure the DataFrame contains the specified columns
joined_df = joined_df[columns]
 
# One-hot encode categorical features
categorical_features = ['category', 'incident_severity', 'road_manager_type', 'road_number', 
                        'road_name', 'place_name', 'municipality_name', 'road_manager_name']
joined_df_encoded = pd.get_dummies(joined_df, columns=categorical_features)
 
# Filter out non-numerical columns early
numerical_cols = joined_df_encoded.select_dtypes(include=['int64', 'float64']).columns
numerical_combinations = combinations(numerical_cols, 3)
 
# Initialize a list to store correlation results
correlation_results = []
 
# Function to calculate correlations for a batch of combinations
def calculate_correlations(batch):
    batch_results = []
    for combo in batch:
        # Calculate the pairwise correlations and sum the absolute values
        corr_1_2 = np.corrcoef(joined_df_encoded[combo[0]], joined_df_encoded[combo[1]])[0, 1]
        corr_1_3 = np.corrcoef(joined_df_encoded[combo[0]], joined_df_encoded[combo[2]])[0, 1]
        corr_2_3 = np.corrcoef(joined_df_encoded[combo[1]], joined_df_encoded[combo[2]])[0, 1]
        correlation_sum = abs(corr_1_2) + abs(corr_1_3) + abs(corr_2_3)
        # Append the correlation data to the batch_results list
        batch_results.append({'Feature1': combo[0], 'Feature2': combo[1], 'Feature3': combo[2], 'Correlation_Sum': correlation_sum})
    return batch_results
 
# Process combinations in batches
batch_size = 1000
batch = []
for i, combo in enumerate(numerical_combinations):
    batch.append(combo)
    if len(batch) == batch_size:
        correlation_results.extend(calculate_correlations(batch))
        batch = []
 
# Process any remaining combinations
if batch:
    correlation_results.extend(calculate_correlations(batch))
 
# Convert the list of dictionaries to a DataFrame
correlation_results = pd.DataFrame(correlation_results)
 
# Sort the correlation results by correlation sum
correlation_results = correlation_results.sort_values(by='Correlation_Sum', ascending=False)
 
# Display top 10 combinations
print(correlation_results.head(10))
 
# Visualize the top combinations
plt.figure(figsize=(14, 8))
top_combinations = correlation_results.head(10)
top_combinations['Features'] = top_combinations.apply(lambda row: f"{row['Feature1']}, {row['Feature2']}, {row['Feature3']}", axis=1)
sns.barplot(x='Correlation_Sum', y='Features', data=top_combinations)
plt.title('Top Combinations of Features with Highest Correlation Sum')
plt.xlabel('Correlation Sum')
plt.ylabel('Feature Combinations')
plt.show()
 # Note in this section where you need to make the changes 
 # here you do not need to change anything 
# Convert timestamp to datetime
joined_df['event_start'] = pd.to_datetime(joined_df['event_start'])
joined_df['event_end'] = pd.to_datetime(joined_df['event_end'])
 
# Extracting date and time features from the start time
joined_df['hour'] = joined_df['event_start'].dt.hour
joined_df['day_of_week'] = joined_df['event_start'].dt.dayofweek
joined_df['month'] = joined_df['event_start'].dt.month
joined_df['year'] = joined_df['event_start'].dt.year
# Here you need to  change based on the analysis you got 
def encode_risk_level(value):
    mapping = {
        'No danger': 6,
        'Minor accident risk': 5,
        'Moderate accident risk': 0,
        'Severe accident risk': [1,2],
        'Extreme danger':[3,4]
    }
    return mapping.get(value, -1)  # Default to -1 if value not found
 
def get_accident_risk_month(month):
    if month in [5, 6]:
        return 'Extreme danger'
    elif month in [7, 10, 11]:
        return 'Severe accident risk'
    elif month in [8, 9, 12]:
        return 'Moderate accident risk'
    elif month in [1, 2]:
        return 'Minor accident risk'
    else:
        return 'No danger'
 
joined_df['accident_risk_month'] = joined_df['month'].apply(get_accident_risk_month)
joined_df['accident_risk_month'] = joined_df['accident_risk_month'].apply(encode_risk_level)


 # Categorize by category and encode
def categorize_by_category(category):
    category = category.lower()
    if 'accelerating' in category:
        return 'Minor accident risk'
    elif 'braking' in category:
        return 'Moderate accident risk'
    elif 'speed' in category:
        return 'Severe accident risk'
    elif 'harsh cornering' in category:
        return 'Extreme danger'
    else:
        return 'No danger'
 
joined_df['accident_risk_category'] = joined_df['category'].apply(categorize_by_category)
joined_df['accident_risk_category'] = joined_df['accident_risk_category'].apply(encode_risk_level)

# Categorize by severity and encode
def categorize_by_severity(severity):
    if 'HC1' in severity or 'SP1' in severity:
        return 'Extreme danger'
    elif 'HC2' in severity or 'SP2' in severity:
        return 'Severe accident risk'
    elif 'HB1' in severity or 'HA1' in severity:
        return 'Moderate accident risk'
    elif 'HB2' in severity or 'HA2' in severity:
        return 'Minor accident risk'
    else:
        return 'No danger'
 
joined_df['accident_risk_severity'] = joined_df['incident_severity'].apply(categorize_by_severity)
joined_df['accident_risk_severity'] = joined_df['accident_risk_severity'].apply(encode_risk_level) 

#Here also 
joined_df['accident_risk_day'] = joined_df['day_of_week'].apply(lambda x: 
    'Extreme danger' if x in [5, 6] else
    'Severe accident risk' if x in [7,10,11] else 
    'Moderate accident risk' if x in [8, 9, 12] else 
    'Minor accident risk' if x in [1,2] else 
    'No danger')
joined_df['accident_risk_day'] = joined_df['accident_risk_day'].apply(encode_risk_level)

# here also # Categorize by speed and encode
joined_df['accident_risk_speed'] = joined_df['speed_kmh'].apply(lambda x: 
    'Extreme danger' if x >= 80 else 
    'Severe accident risk' if 60 <= x < 80 else 
    'Moderate accident risk' if 40 <= x < 60 else 
    'Minor accident risk' if 20 <= x < 40 else 
    'No danger')
joined_df['accident_risk_speed'] = joined_df['accident_risk_speed'].apply(encode_risk_level)





# Assign overall risk category and encode
def assign_overall_risk_category(row):
    categories = [row['accident_risk_month'], row['accident_risk_category'], 
                  row['accident_risk_severity'], row['accident_risk_day'], row['accident_risk_speed']]
    if 4 in categories:
        return 1.0  # Extreme danger
    elif 3 in categories:
        return 0.8  # Severe accident risk
    elif 2 in categories:
        return 0.6  # Moderate accident risk
    elif 1 in categories:
        return 0.4  # Minor accident risk
    else:
        return 0.2  # No danger
 
joined_df['risk_category'] = joined_df.apply(assign_overall_risk_category, axis=1)

# Creating behavioral features based on safe driving category
joined_df['harsh_cornering'] = joined_df['category'].apply(lambda x: 1 if 'harsh cornering' in x.lower() else 0)
joined_df['speeding'] = joined_df['category'].apply(lambda x: 1 if 'speeding' in x.lower() else 0)
joined_df['accelerating'] = joined_df['category'].apply(lambda x: 1 if 'accelerating' in x.lower() else 0)
joined_df['braking'] = joined_df['category'].apply(lambda x: 1 if 'braking' in x.lower() else 0)
 
# Including speed, duration, and maxwaarde features
joined_df['speed_kmh'] = joined_df['speed_kmh']
joined_df['end_speed_kmh'] = joined_df['end_speed_kmh']
joined_df['duration_seconds'] = joined_df['duration_seconds']
joined_df['maxwaarde'] = joined_df['maxwaarde']
 
# Creating interaction features
joined_df['speed_maxwaarde_interaction'] = joined_df['speed_kmh'] * joined_df['maxwaarde']
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt

# Define continuous features
continuous_features = ['speed_kmh', 'end_speed_kmh', 'duration_seconds', 'maxwaarde', 'speed_maxwaarde_interaction', 'hour', 'month', 'year']

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the continuous features
joined_df[continuous_features] = scaler.fit_transform(joined_df[continuous_features])

# Check for non-numeric columns in X
non_numeric_columns = joined_df.select_dtypes(include=['object', 'bool']).columns
print(f"Non-numeric columns to be encoded: {non_numeric_columns}")

# Encode non-numeric columns
label_encoder = LabelEncoder()
for column in non_numeric_columns:
    joined_df[column] = label_encoder.fit_transform(joined_df[column].astype(str))

# Ensure target variables are categorical
joined_df['risk_category'] = joined_df['risk_category'].astype(str)
joined_df['incident_severity'] = joined_df['incident_severity'].astype(str)

# Combine risk_category and incident_severity into a single target variable
joined_df['combined_target'] = joined_df['risk_category'] + '_' + joined_df['incident_severity']

# Prepare features and target
X = joined_df.drop(['risk_category', 'incident_severity', 'combined_target', 'event_start', 'event_end', 'category', 'is_valid', 'accident_risk_category', 'accident_risk_month', 'accident_risk_severity', 'accident_risk_day', 'accident_risk_speed'], axis=1)
y = joined_df['combined_target']

# Check the unique classes in the target variable
print("Unique classes in target variable before SMOTE:", np.unique(y))
print("Class distribution before SMOTE:")
print(y.value_counts())

# Remove classes with fewer than 5 samples
min_samples = 5
class_counts = y.value_counts()
valid_classes = class_counts[class_counts >= min_samples].index
X = X[y.isin(valid_classes)]
y = y[y.isin(valid_classes)]

# Check the unique classes in the target variable after filtering
print("Unique classes in target variable after filtering:", np.unique(y))
print("Class distribution after filtering:")
print(y.value_counts())

# Handle class imbalance using SMOTE with adjusted k_neighbors
smote = SMOTE(random_state=42, k_neighbors=1)  # Adjust k_neighbors based on the minimum number of samples in any class
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check the unique classes in the target variable after SMOTE
print("Unique classes in target variable after SMOTE:", np.unique(y_resampled))
print("Class distribution after SMOTE:")
print(pd.Series(y_resampled).value_counts())

# Split the resampled data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Check the unique classes in the target variable after splitting
print("Unique classes in training set:", np.unique(y_train))
print("Unique classes in test set:", np.unique(y_test))

# Initialize the classifiers
models = {
    "Logistic Regression": LogisticRegression(max_iter=250, random_state=42),
    
}

# Train and the model
for model_name, model in models.items():
    print(f"\n--- {model_name} ---")
    # Train the model
    model.fit(X_train, y_train)
    # Predict on the test set
    y_pred = model.predict(X_test)
    # Evaluate the model
    print(classification_report(y_test, y_pred))
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
  
```
